#1. Análisis de facturación



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Identificamos la facturacion
Ingresos_tienda = tienda["Precio"].sum()
Ingresos_tienda2 = tienda2["Precio"].sum()
Ingresos_tienda3 = tienda3["Precio"].sum()
Ingresos_tienda4 = tienda4["Precio"].sum()

df_ingresos = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Ingresos": [Ingresos_tienda, Ingresos_tienda2, Ingresos_tienda3, Ingresos_tienda4]
}).set_index("Tienda")
```

```python
ax = df_ingresos.plot(kind="bar", figsize=(8, 4), color="blue")

ax.set_xlabel("Ingresos")
ax.set_title("Ingresos por tienda")
ax.bar_label(ax.containers[0])

plt.show()
```

```python
df_ingresos_mill = df_ingresos / 1_000_000

ax = df_ingresos_mill.plot(kind="bar", figsize=(8, 4), color="pink")
ax.set_ylabel("Ingresos (millones)")
ax.set_title("Ingresos por tienda")

ax.bar_label(ax.containers[0], fmt="%.2f")  # valores ya en millones

plt.show()
```

# 2. Ventas por categoría

```python
#Hallamos la cantidad de ventas por cada categoria de la tienda
tienda.groupby('Categoría del Producto')["Precio"].count()
```

```python
ventas_cat = (
    tienda
    .groupby('Categoría del Producto')["Precio"]
    .count()
    .sort_values(ascending=False)
)

print(ventas_cat)
```

```python
tiendas = {
    'tienda1': tienda,
    'tienda2': tienda2,
    'tienda3': tienda3,
    'tienda4': tienda4
}

resultados = {}

for nombre, df in tiendas.items():
    res = (
        df
        .groupby('Categoría del Producto')
        .agg(Cantidad_ventas=('Precio', 'count'))
        .sort_values('Cantidad_ventas', ascending=False)
    )
    resultados[nombre] = res
    print(f'\n=== {nombre} ===')
    print(res)
```

```python
res_list = []
for nombre, df in tiendas.items():
    tmp = (
        df
        .groupby('Categoría del Producto')
        .agg(Cantidad_ventas=('Precio', 'count'))
        .rename(columns={'Cantidad_ventas': nombre})
    )
    res_list.append(tmp)

comparacion = pd.concat(res_list, axis=1).fillna(0).astype(int)
print(comparacion)
```

```python
comparacion['Total'] = comparacion.sum(axis=1)
comparacion = comparacion.sort_values('Total', ascending=False)
print(comparacion)
```

# 4. Productos más y menos vendidos

```python
p_tienda1 = tienda.groupby("Producto")["Precio"].count().sort_values(ascending=False)
p_tienda1.head(5)
```

```python
p_tienda1.tail(5)
```

```python
def top_y_bottom(df, nombre_tienda, n=5):
    conteo = (
        df.groupby("Producto")["Precio"]
          .count()
          .sort_values(ascending=False)
          .rename("Cantidad_ventas")
    )

    top_n = conteo.head(n)
    bottom_n = conteo.tail(n)

    # Los paso a DataFrame y agrego la columna Tienda para poder unir después
    top_n = top_n.reset_index()
    top_n["Tienda"] = nombre_tienda
    top_n["Tipo"] = "Más vendidos"

    bottom_n = bottom_n.reset_index()
    bottom_n["Tienda"] = nombre_tienda
    bottom_n["Tipo"] = "Menos vendidos"

    return top_n, bottom_n
```

```python
tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

lista_res = []

for nombre, df in tiendas.items():
    top_n, bottom_n = top_y_bottom(df, nombre, n=5)
    lista_res.append(top_n)
    lista_res.append(bottom_n)

resumen_productos = pd.concat(lista_res, ignore_index=True)
```

# 5. Envío promedio por tienda

```python
epro_t1 = tienda['Costo de envío'].mean()
epro_t2 = tienda2['Costo de envío'].mean()
epro_t3 = tienda3['Costo de envío'].mean()
epro_t4 = tienda4['Costo de envío'].mean()
df_epro = pd.DataFrame({
    "Tienda": ["Tienda 1","Tienda 2","Tienda 3","Tienda 4"],
    "Costo de envío": [epro_t1,epro_t2,epro_t3,epro_t4]
}).set_index("Tienda")
df_epro['Costo de envío'] = df_epro['Costo de envío'].round(0)
```

```python
px = df_epro.plot(kind= "bar",figsize = (8,4), color = "Green")
px.set_xlabel("Costo promedio de envio")
px.set_title("Costo promedio de envio por tienda")
px.bar_label(px.containers[0], padding=-10, color = "white")
plt.show()
```
