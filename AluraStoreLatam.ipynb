#1. Análisis de facturación



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Identificamos la facturacion
Ingresos_tienda = tienda["Precio"].sum()
Ingresos_tienda2 = tienda2["Precio"].sum()
Ingresos_tienda3 = tienda3["Precio"].sum()
Ingresos_tienda4 = tienda4["Precio"].sum()

df_ingresos = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Ingresos": [Ingresos_tienda, Ingresos_tienda2, Ingresos_tienda3, Ingresos_tienda4]
}).set_index("Tienda")
```

```python
ax = df_ingresos.plot(kind="bar", figsize=(8, 4), color="blue")

ax.set_xlabel("Ingresos")
ax.set_title("Ingresos por tienda")
ax.bar_label(ax.containers[0])

plt.show()
```

```python
df_ingresos_mill = df_ingresos / 1_000_000

ax = df_ingresos_mill.plot(kind="bar", figsize=(8, 4), color="pink")
ax.set_ylabel("Ingresos (millones)")
ax.set_title("Ingresos por tienda")

ax.bar_label(ax.containers[0], fmt="%.2f")  # valores ya en millones

plt.show()
```

# 2. Ventas por categoría

```python
#Hallamos la cantidad de ventas por cada categoria de la tienda
tienda.groupby('Categoría del Producto')["Precio"].count()
```

```python
ventas_cat = (
    tienda
    .groupby('Categoría del Producto')["Precio"]
    .count()
    .sort_values(ascending=False)
)

print(ventas_cat)
```

```python
tiendas = {
    'tienda1': tienda,
    'tienda2': tienda2,
    'tienda3': tienda3,
    'tienda4': tienda4
}

resultados = {}

for nombre, df in tiendas.items():
    res = (
        df
        .groupby('Categoría del Producto')
        .agg(Cantidad_ventas=('Precio', 'count'))
        .sort_values('Cantidad_ventas', ascending=False)
    )
    resultados[nombre] = res
    print(f'\n=== {nombre} ===')
    print(res)
```

```python
res_list = []
for nombre, df in tiendas.items():
    tmp = (
        df
        .groupby('Categoría del Producto')
        .agg(Cantidad_ventas=('Precio', 'count'))
        .rename(columns={'Cantidad_ventas': nombre})
    )
    res_list.append(tmp)

comparacion = pd.concat(res_list, axis=1).fillna(0).astype(int)
print(comparacion)
```

```python
comparacion['Total'] = comparacion.sum(axis=1)
comparacion = comparacion.sort_values('Total', ascending=False)
print(comparacion)
```

```python
ventas1 = tienda.groupby("Categoría del Producto").size().rename("Tienda 1")
ventas2 = tienda2.groupby("Categoría del Producto").size().rename("Tienda 2")
ventas3 = tienda3.groupby("Categoría del Producto").size().rename("Tienda 3")
ventas4 = tienda4.groupby("Categoría del Producto").size().rename("Tienda 4")
```

```python
df_ventas_categorias = pd.concat([ventas1, ventas2, ventas3, ventas4],axis=1).fillna(0).astype(int)
df_ventas_categorias.head()
```

```python
##Comparacion de las categorias mas vendidas por tienda
df_ventas_categorias["Total"] = df_ventas_categorias.sum(axis=1)
df_ventas_categorias = df_ventas_categorias.sort_values("Total", ascending=False)
df_ventas_categorias
```
